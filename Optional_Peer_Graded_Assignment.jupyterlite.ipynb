#### Read the dataset in the csv file from the URL


```python
from js import fetch
import io

URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ST0151EN-SkillsNetwork/labs/boston_housing.csv'
resp = await fetch(URL)
boston_url = io.BytesIO((await resp.arrayBuffer()).to_py())
```

```python
boston_df=pd.read_csv(boston_url)
```

#### Add your code below following the instructions given in the course to complete the peer graded assignment


```python
plt.figure(figsize=(10, 6))
sns.histplot(boston_df['MEDV'], bins=20, kde=True)
plt.title('Гистограмма Медианной стоимости домов')
plt.xlabel('Медианная стоимость домов, $')
plt.ylabel('Частота')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='CHAS', data=boston_df)
plt.title('Гистограмма для переменной Charles River')
plt.xlabel('Расположение у реки Charles (1 - да, 0 - нет)')
plt.ylabel('Частота')
plt.show()
```

```python
boston_df['AGE_group'] = pd.cut(boston_df['AGE'], bins=[0, 35, 70, np.inf], labels=['35 лет и моложе', 'от 35 до 70 лет', '70 лет и старше'])

plt.figure(figsize=(10, 6))
sns.histplot(data=boston_df, x='MEDV', hue='AGE_group', bins=20, kde=True, multiple='stack')
plt.title('Гистограмма MEDV по возрастным группам')
plt.xlabel('Медианная стоимость домов, $')
plt.ylabel('Частота')
plt.legend(title='Возрастные группы')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='NOX', y='DIS', data=boston_df)
plt.title('Диаграмма рассеяния для NOX и DIS')
plt.xlabel('Концентрация оксида азота, ppm')
plt.ylabel('Расстояние до центров занятости, миль')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(boston_df['PTRATIO'], bins=20, kde=True)
plt.title('Гистограмма для PTRATIO')
plt.xlabel('Соотношение учеников и учителей')
plt.ylabel('Частота')
plt.show()
```
